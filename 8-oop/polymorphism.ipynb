```python
## polymorphism
## overriding

# base class
class animal:
    def speak(self):
        return 'sound of the animal'

# derived class
class dog(animal):
    def speak(self):
        return 'woof'
    
# derived class 2
class cat(animal):
    def speak(self):
        return 'meow'

cat1 = cat()
dog1 = dog()
print(cat1.speak())
print(dog1.speak())
```

```python
from math import pi
class shape:
    def area(self):
        return "area of the shape"
    
class rectangle(shape):
    def __init__(self,width,length):
        self.width = width
        self.length = length
        
    def area(self):
        return self.length * self.width
    
class circle(shape):
    def __init__(self,radius):
        self.radius = radius
    def area(self):
        return pi * self.radius * self.radius
    

def areaa(shape):
    return shape.area()

rectangle1 = rectangle(4,5)
circle1 = circle(3)
print(areaa(rectangle1))
print(areaa(circle1))
    
```

```python
## polymorphism with abstract base classes

from abc import ABC,abstractclassmethod

class vehicle(ABC):
    @abstractclassmethod
    def start_engine(self):
        pass

class car(vehicle):
    def start_engine(self):
        return "car started"
    
class motorcycle(vehicle):
    def start_engine(self):
        return "motorcycle started"
    
def start(vehicle):
    return vehicle.start_engine()

car1 = car()
motorcycle1 = motorcycle()
print(car1.start_engine())
print(motorcycle1.start_engine())
```
