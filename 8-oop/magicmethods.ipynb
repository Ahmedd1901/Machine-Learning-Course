```python
## magic methods
class person:
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name} is {self.age} years old"
    def __repr__(self):
        return f"{self.name} is {self.age} years old"
person1 = person("ahmed",24)
print(person1)
print(repr(person1))
```
