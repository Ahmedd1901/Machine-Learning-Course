```python
## Read a whole file

with open('example.txt','r') as file:
    content = file.read()
    print(content)
```

```python
## read a file line by line
with open('example.txt','r') as file:
    for line in file:
        print(line)
```

```python
## writing a file
with open('example2.txt','w') as file:
    file.write("Hello world\n")
    file.write("Hiiiii")
```

```python
## writing a file (overwriting)
with open('example2.txt','w') as file:
    file.write('hellooo\n')
```

```python
## writing a file (appending)
with open('example2.txt','a') as file:
    file.write('ahmed\n')
```

```python
## writing a list of lines to a file
lines = ['first line\n','second line\n','third line\n']
with open('example2.txt','a') as file:
    file.writelines(lines)
```

```python
## binary files
data = b'\x00,\x01,\x02'
with open('example.bin','wb') as file:
    file.write(data)
```

```python
with open('example.bin','rb') as file:
    content = file.read()
    print(content)
```

```python
## read from a source file and write to destination file
with open('example.txt','r') as file:
    content = file.read()

with open('destination.txt','w') as file:
    file.write(content)
```

```python
## read a text file and count number of lines, words and characters.

def count_text_file(file_path):
    with open(file_path,'r') as file:
        lines = file.readlines()
        line_count = len(lines)
        words_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    return line_count,words_count,char_count

print(count_text_file('example.txt'))
```

```python
## writing and then reading
with open('source.txt','w+') as file:
    file.write("hi, how are you?\n")
    file.write("hope you are doing well\n")

    ## move cursor to the beginning 
    file.seek(0)
    content = file.read()
    print(content)
```
