```python
# ### Assignment 6: List Methods

# Create a list of random numbers and sort 
# it in ascending and descending order. Remove the duplicates 
# from the list and print the modified list.

import random

random_list = [random.randint(1,10) for i in range(10)]
print(f"random list is {random_list}")
sorted_asc = sorted(random_list)
print(f"the list sorted ascendingly is {sorted_asc}")
sorted_desc = sorted(random_list,reverse=True)
print(f"the list sorted descendingly is {sorted_desc}")
modified_list = set(random_list)
print(f"the list without duplicates: {modified_list}")

```

```python
# ### Assignment 13: List Reversal

# Write a function that takes a list and returns 
# a new list with the elements in reverse order. 
# Print the original and reversed lists.
def rev(listt):
    return listt[::-1]
newL = [1,2,3,4,5]
print (newL)
print(rev(newL))
# ### Assignment 14: List Rotation

# Write a function that rotates a list by n positions. 
# Print the original and rotated lists.
def rotate_list(lst,n):
    for i in range(n):
        last = lst.pop()
        lst.insert(0,last)
    return lst
listtt = [1,2,3,4,5]
print(listtt)
print(rotate_list(listtt,2))
# ### Assignment 15: List Intersection

# Write a function that takes two lists and 
# returns a new list containing only the elements 
# that are present in both lists. Print the intersected list.
def list_intersection(lst1,lst2):
    lst = [i for i in lst1 if i in lst2]
    return lst
listt1 = [1,2,3,4]
listt2 = [3,4,5,6]
print(list_intersection(listt1,listt2))
```
