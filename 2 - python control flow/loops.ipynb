## Loops

```python
range(5)
```

```python
## for loop
for i in range(5):
    print(i)
```

```python
for i in range (1,4):
    print(i)
```

```python
for i in range (1 , 10 , 2):
    print(i)
```

```python
for i in range(10 , 1, -2):
    print(i)
```

```python
name = "ahmed"
for i in name:
    print(i)
```

```python
## while loop
count = 0 
while count < 3:
    print(count)
    count += 1
```

```python
## Loop control statements
## break, breaks the loop and exits it at certain condition

for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
## continue, skips the current iteration and continues to next
for i in range(10):
    if i == 5:
        continue
    print(i)
```

```python
## pass

for i in range(5):
    if i == 3:
        pass
    print(i)
```

```python
## nested loops

for i in range(3):
    for j in range(2):
        print(f"i is {i} and j is {j}")
```

```python
## prime numbers
for num in range(1,30):
    if num > 1:
        for i in range(2, num):
            if num % i == 0:
                break
        else:
            print(num)

## the else of the for loop runs if the loop finishes naturally without
## any breaks
```
