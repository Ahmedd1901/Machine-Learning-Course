## Conditional Statements


```python
## if statement
age = 24
if(age >= 20):
    print("accepted")

## else statement
x = 5
if(x > 6):
    print("allowed")
else:
    print("not allowed")

## elif statement
x = 11
if(10>x>5):
    print("greater than 5")
elif (x>10):
    print("greater than 10")
else:
    print("ok")
```

```python
## nested if
num = int(input("enter a number"))

if(num > 0):
    print("the number is positive")
    if(num%2 == 0):
        print("the number is even")
    else:
        print("the number is odd")
```

```python
## leap year example
year = int(input("enter the year"))

if(year % 4 == 0 ):
    if(year % 100 == 0):
        if(year % 400 == 0):
            print("it's a leap year")
        else:
            print("it's not a leap year")
    else:
        print("it's a leap year")
else:
    print("it's not a leap year")
    
```

```python
## simple calculator
num1 = float(input("enter first number"))
num2 = float(input("enter second number"))
operation = input("enter operation(+,-,*,/)")

if(operation == '+'):
    result = num1 + num2
elif(operation == '-'):
    result = num1 - num2
elif(operation == '*'):
    result = num1 * num2
elif(operation == '/'):
    result = num1 / num2
else:
    print("invalid operation")

print(result)
```
