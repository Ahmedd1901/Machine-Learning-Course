### Pandas-Dataframe and series

```python
import pandas as pd
## series
data = [1,2,3,4,5]
series = pd.Series(data)
print(series)
```

```python
## create a series from dictionary
data = {'a': 1 , 'b':2, 'c':3 , 'd': 4}
series_dict = pd.Series(data)
print(series_dict)
```

```python
data2 = [10, 20, 30]
indexes = ['a','b','c']
print(pd.Series(data2,index= indexes))
```

```python
## DataFrame
## Create a DataFrame from a dictionary of list

data3 = {
    'Name': ['ahmed','mohamed','yousef'],
    'Age': [25,21,22],
    'City':['cairo','alexandria','hurghada']
}
df = pd.DataFrame(data3)
print(df)
```

```python
import numpy as np
np.array(df)
```
