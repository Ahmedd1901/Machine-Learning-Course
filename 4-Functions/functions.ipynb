```python
def even_or_odd(number):
    """THis function finds if the number is odd or even"""
    if number % 2 == 0:
        print("the number is even")
    else:
        print("the number is odd")
```

```python
## function with multiple parameters
def add(a,b):
    return a + b

result = add(2,4)
print(result)
```

```python
## default parameters

def greet(name = "Noname"):
    print(f"Hello {name}")

greet("ahmed")
greet()
```

```python
### variable length arguments
## positional and keywords arguments

## positional arguments
def print_numbers(*args):
    for numbers in args:
        print(numbers)


print_numbers(1,2,3,4,5,6,"ahmed")
```

```python
## keywords argument

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")

print_details(name = "ahmed",age = 24,)
```

```python
def print_details2(*args,**kwargs):
    for val in args:
        print(f"positional arguments: {val}")

    for key,value in kwargs.items():
        print(f"keywords arguments: {key}:{value}")


print_details2(1,2,3,name="ahmed",age = 24)
```
