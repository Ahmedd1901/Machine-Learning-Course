## map

```python
def square(x):
    return x**2

square(10)
```

```python
numbers = [1,2,3,4,5,6,7,8]
list(map(square,numbers))
```

```python
# lambda with map 
list(map(lambda x:x*4, numbers))
```

```python
numbers1 = [1,2,3]
numbers2 = [4,5,6]

added_numbers = list(map(lambda x,y:x+y,numbers1,numbers2))
print(added_numbers)
```

```python
str_numbers = ['1','2','3','4','5']
int_numbers = list(map(int, str_numbers))
print(int_numbers)
```

```python
words = ["ahmed","hany"]
upper_words = list(map(str.upper,words))
print(upper_words)
```

```python
def getName(person):
    return person['name']

people = [
    {'name':'ahmed','age':24},
    {'name':'mohamed','age':20}

]

list(map(getName,people))
```
