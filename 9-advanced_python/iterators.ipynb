```python
my_list = [1,2,3,4,5,6]
for i in my_list:
    print(i)
```

```python
## Iterator
iterator = iter(my_list)
print(iterator)
```

```python
next(iterator)
```

```python
next(iterator)
```

```python
try:
    print(next(iterator))
except StopIteration:
    print("There are no more elements")
```
