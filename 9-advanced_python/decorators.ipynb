```python
## closures
def main_welcome(msg):
    msg = "welcome"
    def sub_welcome():
        print(msg)
        print("welcome again")  
    return sub_welcome()
```

```python
main_welcome("hello")
```

```python
def main_welcome(func):
    msg = "welcome"
    def sub_welcome():
        func()
        print(msg)
        print("welcome again")  
    return sub_welcome()
```

```python
main_welcome(print)
```

```python
## decorator
@main_welcome
def func():
    print("hiii")
```

```python
def repeat(n):
    def decorator(func):
        def wrapper(*args,**kwargs):
            for _ in range(n):
                func(*args,**kwargs)
        return wrapper
    return decorator
```

```python
@repeat(3)
def hello():
    print("hello")
```

```python
hello()
```
