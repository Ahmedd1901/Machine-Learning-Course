```python
def square(n):
    for i in range(n):
        return i**2
```

```python
square(3)
```

```python
def squaree(n):
    for i in range(n):
        yield i**2
    
```

```python
squaree(3)
```

```python
for i in squaree(4):
    print(i)
```

```python
## practical example
## reading large files
def read_large_files(file_path):
    with open(file_path,'r') as file:
        for line in file:
            yield line.strip()

gen = read_large_files(file_path="large_file.txt")
```
