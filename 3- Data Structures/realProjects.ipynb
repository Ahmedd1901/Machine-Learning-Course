## To Do list

```python
to_do_list = ["buy groceries","clean the house","pay bills"]

## adding a task
to_do_list.append("go to gym")

## remove a task
to_do_list.remove("pay bills")

## check if a task is in the list
if "buy groceries" in to_do_list:
    print("don't forget the tomato")

print("remaining tasks: ")
for task in to_do_list:
    print(f"-{task}")
```

## student grades

```python
grades = [85,92,78,95,83]

## adding a new grade
grades.append(89)

## finding average of grades
avg_grade = sum(grades) / len(grades)
print(f"the average grade is: {avg_grade}")

## finding heighest and lowest grade
print(f"the highest grade is: {max(grades)}")
print(f"the lowest grade is: {min(grades)}")
```
