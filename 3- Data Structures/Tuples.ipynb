## Tuples


```python
empty_tuple =()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst = list()
print(type(lst))
tpl = tuple()
print(type(tpl))
```

```python
numbers = tuple([1,2,3,4,5,6])
print(numbers)
mixed_tuple = (1,"ahmed",True,5.4)
print(mixed_tuple)
print(type(mixed_tuple))
```

```python
print(numbers[-1])
print(numbers[0:4])
print(numbers[::-1])
```

```python
concatenated_tuple = numbers + mixed_tuple
print(concatenated_tuple)
print(mixed_tuple * 2)
```

```python
## Immutable nature of tuples
lst = [1,2,3,4,5]
print(lst)
lst[1] = "krish"
print(lst)

tuplee = (1,2,3)
# tuplee[1] = "krish"  unsupported
```

```python
## tuple methods
print(numbers.count(1)) # how many 1's in the tuple
print(numbers.index(3)) # at which index is 3
```

```python
## packing a tuple
Packed_tuple = 1,"hello", 2.5
print(Packed_tuple)

## unpacking tuple
a,b,c = Packed_tuple
print(a)
print(b)
print(c)
```

```python
## unpacking with *
numberss = (1,2,3,4,5,6,7)
start,middle , *end = numberss
print(start)
print(middle)
print(end)
```

```python
## nested tuple
nested_tuple = ((1,2,3), ("a","b"),(True,False))
print(nested_tuple[0][2])
```

```python
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")
    print()  
```
