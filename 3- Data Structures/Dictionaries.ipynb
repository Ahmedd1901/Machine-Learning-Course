## Dictionaries


```python
empty_dict = dict()
empty_dict2 = {}
print(type(empty_dict))
print(type(empty_dict2))
```

```python
student = {"name": "ahmed", "age": 24 , "grade":"A"}
print(student)
print(student["age"])
print(student.get("name"))
student["address"] = "egypt" ## add key and its value
print(student)
```

```python
del student['grade'] ## delete key and its value
print(student)
```

```python
keys = student.keys() ## get the keys only
values = student.values() ## get the values only
print(keys)
print(values)
print(type(keys))
items = student.items() ## get the key value pairs
print(items)
print(type(items))
```

```python
## shallow copy
student_copy = student
print(student)
print(student_copy)
```

```python
student["name"] = "ahmed2"
print(student)
print(student_copy)
```

```python
student_copy2 = student.copy()
print(student_copy2)
print(student)
```

```python
student["age"] = 25
print(student_copy2)
print(student)
```

```python
## iterating over dictionaries
for keyy in student.keys():
    print(keyy)
```

```python
for valuee in student.values():
    print(valuee)
```

```python
for itemm in student.items():
    print(itemm)

for key,value in student.items():
    print(f"{key}:{value}")
```

```python
students = {
    "student1":{"name":"ahmed", "age":25},
    "student2":{"name":"mohamed", "age":21},
}

print(students["student1"]["name"])
print(students["student2"]["age"])
```

```python
for student_id,student_info in students.items():
    print(f"{student_id}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
## dictionary comprehension
squares = {x:x**2 for x in range(5)}
print(squares)
print(type(squares))
```

```python
## use a dictionary to count the frequency of elements in list

numbers = [1,2,2,3,4,4,4,5,6,7,7,7]
frequency = {}
for number in numbers:
    if number in frequency:
        frequency[number] += 1
    else:
        frequency[number] = 1
print(frequency)
```

```python
## merge two dictionaries into one 
dict1 = {"a" :1,"b": 2}
dict2 = {"c" :3,"d": 4}
merged_dict = {**dict1,**dict2}
print(merged_dict)
```
