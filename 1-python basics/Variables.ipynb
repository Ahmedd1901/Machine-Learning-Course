## Variables In Python


```python
## declaring and assigning variables
## variables should start with a letter or an underscore 
## variables can contain letters, numbers, and underscores 
age = 24
name = "ahmed"
height = 178.5
is_student = False

print("age : ",age)
print("name : ",name)
print("height : ",height)
print("student : ",is_student)
```

```python
## type of a variable is determined at runtime
age = 24 #int
name = "ahmed" #str
print("the type of the variable age is :",type(age))
## type conversion
age = str(age)
print("the type of the variable age after conversion is :",type(age))
```

```python
## input
age = int(input("What is your age?"))
print(age,type(age))
```

```python
## simple calculator
Num1 = float(input("enter first number"))
Num2 = float(input("enter second number"))

sum = Num1 + Num2
difference = Num1 - Num2
product = Num1 * Num2
quotient = Num1 / Num2

print("sum: ",sum)
print("difference: ", difference)
print("product: ",product)
print("Quotient: ",quotient)
```
