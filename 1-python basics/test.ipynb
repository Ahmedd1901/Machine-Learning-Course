## Basic Syntax Rules In Python

```python
## Python is case sensitive
name = "Ahmed"
Name = "Hany"

print(name)
print(Name)
```

```python
## Indentation
## Python uses indentation to define blocks of code
age = 32
if age > 30:
    print(age)
else:
    print(0)
```

```python
## Line Continuation

total = 1 + 2 + 3 + 4 + 5 +\
6 + 7

print(total)
```

```python
## Multiple Statements on a single line

x = 5; y = 6; z = x + y
print(z)
```

```python
## Semantics
# variable assignment
age = 32 
name = "ahmed"
```

```python
type(age)
```

```python
type(name)
```

```python
## Type Inference
var = 10
print(type(var))
var = "ahmed"
print(type(var))
```
