## Data Types


```python
## integer
age = 24
print(type(age))
## float
height = 6.5
print(type(height))
## string
name = "ahmed"
print(type(name))
## boolean
is_student = True
print(type(is_student))

```

## Operators

```python
## Arithmetic Operation

a = 13; b = 5
add_result = a + b
sub_result = a - b
floor_div_result = a // b
modulus_result = a % b
exponential_result = a**b
print(add_result)
print(sub_result)
print(floor_div_result)
print(modulus_result)
print(exponential_result)
```

```python
## comparison operators
## equal
a = 10; b=10
print(a == b)

## not equal
print(a != b)

## greater than , greater than or equal
print(a > b)
print(a >= b)

## less than , less than or equal
print(a < b)
print(a <= b)

```

```python
## Logical Operators
x = True
y = False

## and
print(x and y)

## or
print(x or y)

## not
print(not x)

```
