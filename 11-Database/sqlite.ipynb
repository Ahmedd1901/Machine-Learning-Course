```python
import sqlite3
```

```python
connection = sqlite3.connect("movies.db")
cursor = connection.cursor()
```

```python
cursor.execute('''
CREATE TABLE IF NOT EXISTS movie(id INTEGER PRIMARY KEY,
    name TEXT)
''')
cursor.execute('''
INSERT OR IGNORE INTO movie VALUES(1,"Oppenheimer")
''')
connection.commit()
```

```python
data = cursor.execute('''
SELECT * FROM movie
''')
for i in data:
    print(i)
```

```python
movies = [
    (2,"Beekeeper"),
    (3,"Fast and furious"),
    (4,"GodFather")
]
cursor.executemany('''
INSERT OR IGNORE INTO movie VALUES(?,?)
''',movies)
connection.commit()
```

```python
data = cursor.execute('''
SELECT * FROM movie
''')
for i in data:
    print(i)
```

```python
cursor.execute('''
SELECT * FROM movie
''')
print(cursor.fetchone())
```

```python
cursor.execute('''
SELECT * FROM movie
''')
print(cursor.fetchmany(2))
print(cursor.fetchone())
```

```python
cursor.execute('''
SELECT * FROM movie
''')
data = cursor.fetchall()
print(data)
for row in data:
    print(row)
```

```python
cursor.execute('''
SELECT * FROM movie ORDER BY name DESC
''')
print(cursor.fetchall())
```

```python
cursor.execute('''
SELECT * FROM movie ORDER BY name LIMIT 2
''')
print(cursor.fetchall())
```

```python
cursor.execute('''
SELECT * FROM movie WHERE name != 'Beekeeper' AND id != 3
''')
print(cursor.fetchall())
```

```python
cursor.execute('''
SELECT * FROM movie WHERE name != UPPER('Beekeeper') 
''')
print(cursor.fetchall())
```

```python
cursor.execute('''
SELECT * FROM movie WHERE name LIKE 'oPpenheimer'
''')
print(cursor.fetchall())
```

```python
cursor.execute('''
SELECT * FROM movie WHERE name LIKE '%e%'
''')
print(cursor.fetchall())
```

```python
cursor.execute('''
SELECT * FROM movie WHERE name LIKE '%s'
''')
print(cursor.fetchall())
```

```python
cursor.execute('''
SELECT * FROM movie WHERE name LIKE '_a%'
''')
print(cursor.fetchall())
```
