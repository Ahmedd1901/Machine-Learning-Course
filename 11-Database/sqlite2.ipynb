```python
import sqlite3
```

```python
connection = sqlite3.connect('movies.db')
cursor = connection.cursor()
cursor.execute('''
UPDATE movie SET name = 'Fast and furious 7' where id = 3
''')
cursor.execute('''
SELECT * FROM movie
''')
print(cursor.fetchall())
```

```python
## delete data
cursor.execute('''
DELETE FROM movie WHERE id = 5
''')
connection.commit()
```

```python
cursor.execute('''
SELECT * FROM movie
''')
print(cursor.fetchall())
```

```python
cursor.execute('''
CREATE TABLE IF NOT EXISTS user(id INTEGER PRIMARY KEY,name TEXT)
''')
connection.commit()
```

```python
cursor.execute('''
INSERT OR IGNORE INTO user VALUES(1,'ahmed')
''')
connection.commit()
```

```python
cursor.execute('''
SELECT * FROM user
''')
print(cursor.fetchall())
```

```python
## delete table
cursor.execute('''
DROP TABLE user
''')
connection.commit()
```

```python
cursor.execute('''
CREATE TABLE IF NOT EXISTS users(id INTEGER PRIMARY KEY AUTOINCREMENT
,name TEXT NOT NULL)
''')
connection.commit()
```

```python
cursor.execute('''
INSERT OR IGNORE INTO users(name) VALUES('ahmed')
''')
connection.commit()
```

```python
cursor.execute('''
SELECT * FROM users
''')
print(cursor.fetchall())
```

```python
cursor.execute("""
ALTER TABLE users
ADD COLUMN age INTEGER
""")
connection.commit()
```

```python
cursor.execute('''
SELECT * FROM users
''')
print(cursor.fetchall())
```

```python
cursor.execute('''
UPDATE users SET age = 24 where id = 1
''')
connection.commit()
```

```python
cursor.execute('''
SELECT * FROM users
''')
print(cursor.fetchall())
```
